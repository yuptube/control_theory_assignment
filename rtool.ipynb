## interactive rendition for making the root locus plot

### import modules

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
import ipywidgets as widgets
from IPython.display import display
from root_locus_plot import PlotRootLocus
```

### For the root locus at 1 loop 

```python
K = 1 # K<143
num = [1 * K, 3 * K] if K != 0 else [1]
den = [1, 14, 45, 50 + K, 3 * K]
G = ctrl.TransferFunction(num,den)
system = PlotRootLocus(G, K)
```

### For the root locus at 2 loops

```python

# Define K  and find closed-loop poles
K = 600
Kt = 1 # Kt >0.0717

# Define the open-loop transfer function G(s)H(s)
num = [1* K, 3*K]  # (s + 3)
den = [1, 14, K* Kt + 45, 50+ K+K * 3*Kt ,1800]  # (s+10)(s)(s^2+4s+5) + K(s+3)(kt+1)
G = ctrl.TransferFunction(num,den)
system = PlotRootLocus(G, K, Kt )
```

## Plotting graph for root locus and step response

```python



# Create and use the PlotRootLocus class
system.plot_root_locus()
system.plot_step_response()
```

## Create a plot for maximun damping factor

```python
# Compute and display max damping factor
zeta_max = system.compute_damping_factor()
print(f"Maximum Damping Factor: {zeta_max} at K_T = {Kt}")
system.plot_bode()
```
